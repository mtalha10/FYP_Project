```python
# **Phishing Domain Detection (Data Visualization)**

### The objective of this notebook is to provide relevant data visualizations of the features extracted in the previous notebook

### Data Loading and Cleaning
### Exploratory Data Analysis (EDA):
### Label distribution
### Correlation heatmap
### Histograms
### Categorical feature distribution
```

```python
# Core Libraries
import numpy as np
import pandas as pd

# Visualization Libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Configure Seaborn for consistent styling
sns.set_style("darkgrid")
```

```python
# Load the dataset
try:
    urldata = pd.read_csv("./Url_Processed.csv")
except FileNotFoundError:
    print("Error: File not found. Ensure the file path is correct.")
    exit()

# Drop unnecessary columns
if "Unnamed: 0" in urldata.columns:
    urldata.drop("Unnamed: 0", axis=1, inplace=True)

# Display basic information about the dataset
print("Dataset Info:")
print(urldata.info())

print("\nFirst 5 Rows:")
print(urldata.head())
```

```python
# Display value counts for labels
label_counts = urldata["label"].value_counts()
print("\nLabel Distribution:\n", label_counts)

# Plot the distribution of URL labels
plt.figure(figsize=(10, 6))
sns.countplot(
    x="label",
    data=urldata,
    palette="viridis",
    order=label_counts.index  # Ensures proper label ordering
)
plt.title("Count of URLs", fontsize=18)
plt.xlabel("Type of URLs", fontsize=14)
plt.ylabel("Number of URLs", fontsize=14)
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(12, 8))
corrmat = urldata.corr(numeric_only=True)

sns.heatmap(
    corrmat,
    square=True,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    cbar=True,
    annot_kws={"size": 10},
)
plt.title("Feature Correlation Heatmap", fontsize=16)
plt.show()
```

```python
# Define key features and their display names
hist_features = ["url_length", "hostname_length", "path_length", "fd_length"]
value_names = ["URL Length", "Hostname Length", "Path Length", "FD Length"]

# Plot histograms
for idx, feature in enumerate(hist_features):
    plt.figure(figsize=(10, 6))
    sns.histplot(data=urldata, x=feature, bins=100, hue="label", kde=True, palette="viridis")
    plt.title(f"Distribution of {value_names[idx]}", fontsize=16)
    plt.xlabel(value_names[idx], fontsize=14)
    plt.ylabel("Number of URLs", fontsize=14)
    plt.xlim(0, 150)
    plt.show()
```

```python
# Identify remaining features
remaining_features = [col for col in urldata.columns if col not in hist_features + ["url", "result", "label"]]

# Plot countplots for categorical features
for feature in remaining_features:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=feature, data=urldata, palette="viridis")
    plt.title(f"Distribution of {feature}", fontsize=16)
    plt.xlabel(feature, fontsize=14)
    plt.ylabel("Number of URLs", fontsize=14)
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Sample data if dataset is too large
if len(urldata) > 100000:
    urldata = urldata.sample(100000, random_state=42)
    print(f"Dataset sampled to 100,000 rows.")
```
